```python
import re
from collections import defaultdict

text = """
Machine learning helps computers learn patterns from data. Artificial intelligence is used to build smart systems. 
Natural language processing allows machines to understand human language. Language models predict the next word in a sequence.
"""

N = 3

# ✅ FIXED: use a sequence that exists in data
input_sequence = "natural language"

print("\nSTEP 1: Preprocessing Text\n")

text = text.lower()
text = re.sub(r'[^a-zA-Z\s]', '', text)
words = text.split()

print("Words:")
print(words)

print("\nSTEP 2: Generating N-grams\n")

ngram_counts = defaultdict(lambda: defaultdict(int))

for i in range(len(words) - N + 1):
    context = tuple(words[i:i+N-1])
    next_word = words[i+N-1]

    ngram_counts[context][next_word] += 1
    print(f"{context} → {next_word}")

print("\nSTEP 3: Calculating Probabilities\n")

ngram_probabilities = defaultdict(dict)

for context in ngram_counts:
    total_count = sum(ngram_counts[context].values())
    print(f"\nContext: {context}")

    for next_word in ngram_counts[context]:
        prob = ngram_counts[context][next_word] / total_count
        ngram_probabilities[context][next_word] = prob
        print(f"{next_word} → {ngram_counts[context][next_word]}/{total_count} = {prob:.2f}")

print("\nSTEP 4: Predicting Next Word\n")

input_words = input_sequence.lower().split()
context = tuple(input_words[-(N-1):])

print("Input Sequence:", input_sequence)
print("Context:", context)

# ✅ FIXED: safe prediction + fallback
if context in ngram_probabilities:
    next_word = max(ngram_probabilities[context],
                    key=ngram_probabilities[context].get)
    print("Predicted Next Word:", next_word)
else:
    print("No prediction available for this context")
```
