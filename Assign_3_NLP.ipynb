```python

text = "Natural Language Processing is a field of Artificial Intelligence that focuses on the interaction between humans and computers."

# Convert to lowercase
text = text.lower()
```

```python
print(text)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Sample input text
text = "Natural Language Processing is a very important part of Machine Learning."

# English stopwords
stop_words = set(stopwords.words('english'))

# Step 1: Tokenization
tokens = word_tokenize(text)

# Step 2: Remove stopwords
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print("Tokens:", tokens)
print("After Stopword Removal:", filtered_tokens)
```

```python
from nltk.util import ngrams
from collections import Counter

unigrams = list(ngrams(filtered_tokens, 1))
print("Unigrams:", unigrams)
```

```python
bigrams = list(ngrams(filtered_tokens, 2))
print("Bigrams:", bigrams)
```

```python
trigrams = list(ngrams(filtered_tokens, 3))
print("Trigrams:", trigrams)
```

```python
import nltk
from nltk.util import ngrams
from collections import defaultdict, Counter

# Step 3: Create Trigrams
trigrams = list(ngrams(filtered_tokens, 3))

print("Trigrams:", trigrams)

# Step 4: Build N-gram model (dictionary)
ngram_model = defaultdict(Counter)

for w1, w2, w3 in trigrams:
    ngram_model[(w1, w2)][w3] += 1

print("\nN-gram Model:")
for key, value in ngram_model.items():
    print(key, dict(value))

# Step 5: Predict next word function
def predict_next_word(word1, word2, model):
    if (word1, word2) in model:
        return model[(word1, word2)].most_common(1)[0][0]
    else:
        return "No prediction available"

# Example prediction
word1 = "Natural"
word2 = "Language"

predicted_word = predict_next_word(word1, word2, ngram_model)
print(f"\nPredicted next word after '{word1} {word2}': {predicted_word}")
```
