```python
import nltk

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

texts = [
    "What is Natural Language Processing: The Definitive Guide ... Natural Language Processing (NLP) is a field of Artificial Intelligence (AI) that enables computers to understand, interpret, generate, and interact with human language (text and speech).",
    "By combining linguistics with machine learning, NLP allows applications like chatbots, translation services, and voice assistants (Siri, Alexa) to process complex human communication for tasks like sentiment analysis, information extraction, and automated responses."
]

# POS tagging
for sentence in texts:
    words = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(words)

    for word, tag in pos_tags:
        print(f"{word} --> {tag}")

```
