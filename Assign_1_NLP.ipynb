```python

nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize
s = "NLP enables computers and digital devices to recognize, understand and generate text and speech by combining computational linguistics, the rule-based modeling of human language together with statistical modeling"
word_tokenize(s)
print(word_tokenize(s))
```

```python
s = "NLP enables computers and digital devices to recognize, understand and generate text and speech by combining computational linguistics, the rule-based modeling of human language together with statistical modeling"
from nltk.tokenize import sent_tokenize
sent_tokenize(s)
print(sent_tokenize(s))
```

```python
words = word_tokenize(s)
print(words)
```

```python
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
for word in words:
  print(stemming.stem(word))
```

```python
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
for word in words:
  print(lemma.lemmatize(word))
```

```python
# remove stopword from above process
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
#remove it from the above lemmatization
for word in words:
  if word not in stop_words:
    print(lemma.lemmatize(word))
```
